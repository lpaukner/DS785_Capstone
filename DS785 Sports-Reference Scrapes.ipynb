```python
### MLB Abbreviation Prep ###

# MLB Teams in 1980
mlb_teams_abbrev_1980 = ['BOS','NYY','TOR','BAL','CHW','CLE','DET','KCR','MIN','HOU','SEA','OAK','CAL','TEX',
                         'ATL','PHI','NYM','MON','MIL','STL','CIN','CHC','PIT','SFG','LAD','SDP']

mlb_teams_names_1980 = ['Boston Red Sox', 'New York Yankees', 'Toronto Blue Jays', 'Baltimore Orioles',
                       'Chicago White Sox', 'Cleveland Indians', 'Detroit Tigers', 'Kansas City Royals', 'Minnesota Twins',
                       'Houston Astros', 'Seattle Mariners', 'Oakland Athletics', 'California Angels', 'Texas Rangers',
                       'Atlanta Braves', 'Philadelphia Phillies', 'New York Mets', 'Montreal Expos',
                       'Milwaukee Brewers', 'St. Louis Cardinals', 'Cincinnati Reds', 'Chicago Cubs', 'Pittsburgh Pirates',
                       'San Francisco Giants', 'Los Angeles Dodgers', 'San Diego Padres']

# Marlins and Rockies join MLB
mlb_teams_abbrev_1993 = ['BOS','NYY','TOR','BAL','CHW','CLE','DET','KCR','MIN','HOU','SEA','OAK','CAL','TEX',
                         'ATL','PHI','NYM','FLA','MON','MIL','STL','CIN','CHC','PIT','SFG','LAD','SDP','COL']

mlb_teams_names_1993 = ['Boston Red Sox', 'New York Yankees', 'Toronto Blue Jays', 'Baltimore Orioles',
                       'Chicago White Sox', 'Cleveland Indians', 'Detroit Tigers', 'Kansas City Royals', 'Minnesota Twins',
                       'Houston Astros', 'Seattle Mariners', 'Oakland Athletics', 'California Angels', 'Texas Rangers',
                       'Atlanta Braves', 'Philadelphia Phillies', 'New York Mets', 'Florida Marlins', 'Montreal Expos',
                       'Milwaukee Brewers', 'St. Louis Cardinals', 'Cincinnati Reds', 'Chicago Cubs', 'Pittsburgh Pirates',
                       'San Francisco Giants', 'Los Angeles Dodgers', 'San Diego Padres', 'Colorado Rockies']

# Angels change name
mlb_teams_abbrev_1997 = ['BOS','NYY','TOR','BAL','CHW','CLE','DET','KCR','MIN','HOU','SEA','OAK','ANA','TEX',
                         'ATL','PHI','NYM','FLA','MON','MIL','STL','CIN','CHC','PIT','SFG','LAD','SDP','COL']

mlb_teams_names_1997 = ['Boston Red Sox', 'New York Yankees', 'Toronto Blue Jays', 'Baltimore Orioles',
                       'Chicago White Sox', 'Cleveland Indians', 'Detroit Tigers', 'Kansas City Royals', 'Minnesota Twins',
                       'Houston Astros', 'Seattle Mariners', 'Oakland Athletics', 'Anaheim Angels', 'Texas Rangers',
                       'Atlanta Braves', 'Philadelphia Phillies', 'New York Mets', 'Florida Marlins', 'Montreal Expos',
                       'Milwaukee Brewers', 'St. Louis Cardinals', 'Cincinnati Reds', 'Chicago Cubs', 'Pittsburgh Pirates',
                       'San Francisco Giants', 'Los Angeles Dodgers', 'San Diego Padres', 'Colorado Rockies']

# Rays and Diamondbacks join MLB
mlb_teams_abbrev_1998 = ['TBD','BOS','NYY','TOR','BAL','CHW','CLE','DET','KCR','MIN','HOU','SEA','OAK','ANA','TEX',
                         'ATL','PHI','NYM','FLA','MON','MIL','STL','CIN','CHC','PIT','SFG','LAD','SDP','COL','ARI']

mlb_teams_names_1998 = ['Tampa Bay Devil Rays', 'Boston Red Sox', 'New York Yankees', 'Toronto Blue Jays', 'Baltimore Orioles',
                       'Chicago White Sox', 'Cleveland Indians', 'Detroit Tigers', 'Kansas City Royals', 'Minnesota Twins',
                       'Houston Astros', 'Seattle Mariners', 'Oakland Athletics', 'Anaheim Angels', 'Texas Rangers',
                       'Atlanta Braves', 'Philadelphia Phillies', 'New York Mets', 'Florida Marlins', 'Montreal Expos',
                       'Milwaukee Brewers', 'St. Louis Cardinals', 'Cincinnati Reds', 'Chicago Cubs', 'Pittsburgh Pirates',
                       'San Francisco Giants', 'Los Angeles Dodgers', 'San Diego Padres', 'Colorado Rockies', 'Arizona Diamondbacks']

# Angels change name (again)
mlb_teams_abbrev_2005 = ['TBD','BOS','NYY','TOR','BAL','CHW','CLE','DET','KCR','MIN','HOU','SEA','OAK','LAA','TEX',
                         'ATL','PHI','NYM','FLA','WSN','MIL','STL','CIN','CHC','PIT','SFG','LAD','SDP','COL','ARI']

mlb_teams_names_2005 = ['Tampa Bay Devil Rays', 'Boston Red Sox', 'New York Yankees', 'Toronto Blue Jays', 'Baltimore Orioles',
                       'Chicago White Sox', 'Cleveland Indians', 'Detroit Tigers', 'Kansas City Royals', 'Minnesota Twins',
                       'Houston Astros', 'Seattle Mariners', 'Oakland Athletics', 'Los Angeles Angels of Anaheim', 'Texas Rangers',
                       'Atlanta Braves', 'Philadelphia Phillies', 'New York Mets', 'Florida Marlins', 'Washington Nationals',
                       'Milwaukee Brewers', 'St. Louis Cardinals', 'Cincinnati Reds', 'Chicago Cubs', 'Pittsburgh Pirates',
                       'San Francisco Giants', 'Los Angeles Dodgers', 'San Diego Padres', 'Colorado Rockies', 'Arizona Diamondbacks']

# Rays and Nationals change names
mlb_teams_abbrev_2008 = ['TBR','BOS','NYY','TOR','BAL','CHW','CLE','DET','KCR','MIN','HOU','SEA','OAK','LAA','TEX',
                         'ATL','PHI','NYM','FLA','WSN','MIL','STL','CIN','CHC','PIT','SFG','LAD','SDP','COL','ARI']

mlb_teams_names_2008 = ['Tampa Bay Rays', 'Boston Red Sox', 'New York Yankees', 'Toronto Blue Jays', 'Baltimore Orioles',
                       'Chicago White Sox', 'Cleveland Indians', 'Detroit Tigers', 'Kansas City Royals', 'Minnesota Twins',
                       'Houston Astros', 'Seattle Mariners', 'Oakland Athletics', 'Los Angeles Angels of Anaheim', 'Texas Rangers',
                       'Atlanta Braves', 'Philadelphia Phillies', 'New York Mets', 'Florida Marlins', 'Washington Nationals',
                       'Milwaukee Brewers', 'St. Louis Cardinals', 'Cincinnati Reds', 'Chicago Cubs', 'Pittsburgh Pirates',
                       'San Francisco Giants', 'Los Angeles Dodgers', 'San Diego Padres', 'Colorado Rockies', 'Arizona Diamondbacks']

# Angels change name (sigh...)
mlb_teams_abbrev_2012 = ['TBR','BOS','NYY','TOR','BAL','CHW','CLE','DET','KCR','MIN','HOU','SEA','OAK','LAA','TEX',
                         'ATL','PHI','NYM','MIA','WSN','MIL','STL','CIN','CHC','PIT','SFG','LAD','SDP','COL','ARI']

mlb_teams_names_2012 = ['Tampa Bay Rays', 'Boston Red Sox', 'New York Yankees', 'Toronto Blue Jays', 'Baltimore Orioles',
                       'Chicago White Sox', 'Cleveland Indians', 'Detroit Tigers', 'Kansas City Royals', 'Minnesota Twins',
                       'Houston Astros', 'Seattle Mariners', 'Oakland Athletics', 'Los Angeles Angels', 'Texas Rangers',
                       'Atlanta Braves', 'Philadelphia Phillies', 'New York Mets', 'Miami Marlins', 'Washington Nationals',
                       'Milwaukee Brewers', 'St. Louis Cardinals', 'Cincinnati Reds', 'Chicago Cubs', 'Pittsburgh Pirates',
                       'San Francisco Giants', 'Los Angeles Dodgers', 'San Diego Padres', 'Colorado Rockies', 'Arizona Diamondbacks']


mlb_years = [1980,1993,1997,1998,2005,2008,2012]
mlb_abbrevs = [mlb_teams_abbrev_1980,mlb_teams_abbrev_1993,mlb_teams_abbrev_1997,mlb_teams_abbrev_1998,
               mlb_teams_abbrev_2005,mlb_teams_abbrev_2008,mlb_teams_abbrev_2012]

mlb_year_abbrev_dict = dict(zip(mlb_years,mlb_abbrevs))

team_swap_dict_1980 = dict(zip(mlb_teams_names_1980,mlb_teams_abbrev_1980))
team_swap_dict_1993 = dict(zip(mlb_teams_names_1993,mlb_teams_abbrev_1993))
team_swap_dict_1997 = dict(zip(mlb_teams_names_1997,mlb_teams_abbrev_1997))
team_swap_dict_1998 = dict(zip(mlb_teams_names_1998,mlb_teams_abbrev_1998))
team_swap_dict_2005 = dict(zip(mlb_teams_names_2005,mlb_teams_abbrev_2005))
team_swap_dict_2008 = dict(zip(mlb_teams_names_2008,mlb_teams_abbrev_2008))
team_swap_dict_2012 = dict(zip(mlb_teams_names_2012,mlb_teams_abbrev_2012))

# Create function to select appropriate dict by year
def mlb_abbrev_select(year):
    
    if (year < 1993):
        teams = mlb_year_abbrev_dict[1980]
        swap_dict = team_swap_dict_1980
    elif (year >= 1993) and (year < 1997):
        teams = mlb_year_abbrev_dict[1993]
        swap_dict = team_swap_dict_1993
    elif (year == 1997):
        teams = mlb_year_abbrev_dict[1997]
        swap_dict = team_swap_dict_1997
    elif (year >= 1998) and (year < 2005):
        teams = mlb_year_abbrev_dict[1998]
        swap_dict = team_swap_dict_1998
    elif (year >= 2005) and (year < 2008):
        teams = mlb_year_abbrev_dict[2005]
        swap_dict = team_swap_dict_2005
    elif (year >= 2008) and (year < 2012):
        teams = mlb_year_abbrev_dict[2008]
        swap_dict = team_swap_dict_2008
    elif year >= 2012:
        teams = mlb_year_abbrev_dict[2012]
        swap_dict = team_swap_dict_2012
        
    return(teams, swap_dict)
```

```python
### MLB Scrapes ###

# Game Results #
def season_results_grabber(team, year):
    
    url = 'https://www.baseball-reference.com/teams/' + team + '/' + str(year) + '-schedule-scores.shtml'
    df = pd.read_html(url)[0]
    
    # Columns for Year and Key
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    key = team + str(year)
    keyCol = [key]*len(df.index)
    df['Key'] = keyCol
    
    # Remove unimportant rows   
    df = df[df['Gm#'] != 'Gm#']
    
    return(df)


# Batting Results #
def batting_results_grabber(year,swap_dict):
    
    url = 'https://www.baseball-reference.com/leagues/majors/' + str(year) + '-standard-batting.shtml'
    df = pd.read_html(url)[0]
    
    # Columns for Team, Year, and Key
    
    #replace existing team column with abbreviation
    df['Tm'].replace(swap_dict, inplace=True)
    
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    df['Key'] = df['Tm'] + str(year)
    
    # Remove unimportant rows   
    df = df[df['Tm'] != 'Tm']
    
    return(df)

# Pitching Results #
def pitching_results_grabber(year,swap_dict):
    
    url = 'https://www.baseball-reference.com/leagues/majors/' + str(year) + '-standard-pitching.shtml'
    df = pd.read_html(url)[0]
    
    # Columns for Team, Year, and Key
    
    #replace existing team column with abbreviation
    df['Tm'].replace(swap_dict, inplace=True)
    
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    df['Key'] = df['Tm'] + str(year)
    
    # Remove unimportant rows   
    df = df[df['Tm'] != 'Tm']
    
    return(df)        


# Run scrapes #

years = range(1980,2022)

all_seasons_df = pd.DataFrame()
all_batting_df = pd.DataFrame()
all_pitching_df = pd.DataFrame()


for year in years:
    
    # get appropriate abbreviations
    teams = mlb_abbrev_select(year)[0]
    swap_dict = mlb_abbrev_select(year)[1]
    
    all_batting_df = all_batting_df.append(batting_results_grabber(year, swap_dict))
    all_pitching_df = all_pitching_df.append(pitching_results_grabber(year, swap_dict))
    
    for team in teams:
        all_seasons_df = all_seasons_df.append(season_results_grabber(team, year))

```

```python
### Save MLB Data ###

all_seasons_df.to_csv('MLB_Season_Results_1980to2021.csv')
all_batting_df.to_csv('MLB_Batting_Results_1980to2021.csv')
all_pitching_df.to_csv('MLB_Pitching_Results_1980to2021.csv')
```

```python
### NFL Scrapes ###

def stat_rank_results_grabber(team, year):
        
    url = 'https://www.pro-football-reference.com/teams/' + team + '/' + str(year) + '.htm'
    df = pd.read_html(url)[0]
    
    # Columns for Team, Year, and Key
    teamCol = [team]*len(df.index)
    df['Team'] = teamCol
    
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    key = team + str(year)
    keyCol = [key]*len(df.index)
    df['Key'] = keyCol
    
    return(df)

def season_results_grabber(team, year):
    
    url = 'https://www.pro-football-reference.com/teams/' + team + '/' + str(year) + '.htm'
    df = pd.read_html(url)[1]
    
    # Columns for Team, Year, and Key
    teamCol = [team]*len(df.index)
    df['Team'] = teamCol
    
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    # Replace useless column with key
    key = team + str(year)
    df[df.columns[4]] = key
    
    # Remove duplicate games by removing away teams and loses at neutral sites
    df = df[df[df.columns[8]] != '@']
    df_new = df.drop(df[(df[df.columns[8]] == 'N') & (df[df.columns[5]] == 'L')].index)
    
    # Remove bye weeks
    df_new = df_new[df_new[df_new.columns[9]] != 'Bye Week']
    
    return(df_new)

def conversions_results_grabber(team, year):
    
    url = 'https://www.pro-football-reference.com/teams/' + team + '/' + str(year) + '.htm'
    df = pd.read_html(url)[2]
    
    # Columns for Team, Year, and Key
    teamCol = [team]*len(df.index)
    df['Team'] = teamCol
    
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    key = team + str(year)
    keyCol = [key]*len(df.index)
    df['Key'] = keyCol
    
    
    return(df)

nfl_team_abbrev_2002 = ['buf','nwe','mia','nyj','cin','pit','cle','rav','oti','clt','htx','jax','kan','rai','sdg','den',
                        'dal','phi','was','nyg','gnb','min','chi','det','tam','nor','atl','car','ram','crd','sfo','sea']

nfl_team_abbrev_1999 = ['buf','nwe','mia','nyj','cin','pit','cle','rav','oti','clt','jax','kan','rai','sdg','den',
                        'dal','phi','was','nyg','gnb','min','chi','det','tam','nor','atl','car','ram','crd','sfo','sea']

nfl_team_abbrev_1997 = ['buf','nwe','mia','nyj','cin','pit','rav','oti','clt','jax','kan','rai','sdg','den',
                        'dal','phi','was','nyg','gnb','min','chi','det','tam','nor','atl','car','ram','crd','sfo','sea']
              
years = range(1997,2022)

all_nfl_stat_rank_df = pd.DataFrame()
all_nfl_seasons_df = pd.DataFrame()
all_nfl_conversions_df = pd.DataFrame()

teams = nfl_team_abbrev_1997

for year in years:
    
    if (year >=1999) and (year < 2002):
        teams = nfl_team_abbrev_1999
    elif year >= 2002:
        teams = nfl_team_abbrev_2002
    
    for team in teams:
        
        all_nfl_stat_rank_df = all_nfl_stat_rank_df.append(stat_rank_results_grabber(team, year))
        all_nfl_seasons_df = all_nfl_seasons_df.append(season_results_grabber(team, year))
        if (year >=1999):
            all_nfl_conversions_df = all_nfl_conversions_df.append(conversions_results_grabber(team, year))

```

```python
### NBA Scrapes ###

import requests

# Season Results #
def nba_season_results_grabber(year):
    months = ['october', 'november', 'december', 'january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september']
    
    if year == 2020:
        months = ['october-2019', 'november', 'december', 'january', 'february', 'march', 'july', 'august', 'september', 'october-2020']
  
    year_df = pd.DataFrame()
    
    for month in months:
        url = 'https://www.basketball-reference.com/leagues/NBA_' + str(year) + '_games-' + month + '.html'
        response = requests.get(url)
        if response.status_code == 200:
            # append next month to df
            year_df = year_df.append(pd.read_html(url)[0])
            
            yearCol = [year]*len(year_df.index)
            year_df['Year'] = yearCol
            
    return(year_df)

# Per Game Stats #
def nba_per_game_results_grabber(year):
    url = 'https://www.basketball-reference.com/leagues/NBA_' + str(year) + '.html'
    
    if year >= 2016:
        df = pd.read_html(url)[4]
    else:
        df = pd.read_html(url)[2]
    
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    return(df)

# Per Shooting Stats #
def nba_shooting_results_grabber(year):
    url = 'https://www.basketball-reference.com/leagues/NBA_' + str(year) + '.html'
    
    if year >= 2016:
        df = pd.read_html(url)[11]
    else:
        df = pd.read_html(url)[9]
    
    yearCol = [year]*len(df.index)
    df['Year'] = yearCol
    
    return(df)


years = range(1998,2022)


all_nba_seasons_df = pd.DataFrame()
all_nba_team_per_game_df = pd.DataFrame()
all_nba_team_shooting_df = pd.DataFrame()

for year in years:
    all_nba_seasons_df = all_nba_seasons_df.append(nba_season_results_grabber(year))
    all_nba_team_per_game_df = all_nba_team_per_game_df.append(nba_per_game_results_grabber(year))
    all_nba_team_shooting_df = all_nba_team_shooting_df.append(nba_shooting_results_grabber(year))
```

```python
### Save NBA Data ###

all_nba_seasons_df.to_csv('NBA_Game_Results_1997to2021.csv')
all_nba_team_per_game_df.to_csv('NBA_Team_Per_Game_1997to2021.csv')
# Combine levels into column names
all_nba_team_shooting_df.columns = all_nba_team_shooting_df.columns.map('_'.join).str.strip('_')
all_nba_team_shooting_df.to_csv('NBA_Team_Shooting_1997to2021.csv')
```
